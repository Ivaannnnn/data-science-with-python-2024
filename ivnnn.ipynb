```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
        'Sales': [150, 200, 250, 300]}
df = pd.DataFrame(data)
```

```python
plt.plot(df['Month'], df['Sales'], marker='o')  # Line graph with markers
plt.title('Monthly Sales')                      # Add a title
plt.xlabel('Month')                             # X-axis label
plt.ylabel('Sales')                             # Y-axis label
plt.grid(True)                                  # Add grid lines
plt.show()                                      # Display the graph
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\ivanf\OneDrive\Documents\datafds.csv\marketing_data.csv" 
df = pd.read_csv(file_path)

print("Columns:", df.columns)
print(df.head())

numerical_columns = df.select_dtypes(include='number').columns
if len(numerical_columns) >= 2:
    x_column = numerical_columns[0]
    y_column = numerical_columns[1]

    plt.plot(df[x_column], df[y_column], marker='o')
    plt.title(f"{x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.grid(True)
    plt.show()
else:
    print("No numerical columns found for plotting.")
```
